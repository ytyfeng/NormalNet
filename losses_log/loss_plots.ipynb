```python
import numpy as np
import matplotlib.pyplot as plt

loss1 = np.loadtxt('test_losses_max.txt')
loss2 = np.loadtxt('train_losses_max.txt')
loss3 = np.loadtxt('test_losses_sum.txt')
loss4 = np.loadtxt('train_losses_sum.txt')
loss5 = np.loadtxt('test_losses_globalFeat.txt')
loss6 = np.loadtxt('train_losses_globalFeat.txt')
epochs = np.arange(100)

def plot_losses(epochs, losses, title="Loss wrt number of epochs"):
    plt.plot(epochs, losses)
    plt.title(title)
    plt.xlabel('epochs')
    plt.ylabel('loss')
    plt.show()

def graph_simple_plot(epochs, trainloss, testloss, title="Loss wrt number of epochs"):
    fig, ax = plt.subplots()
    line1, = ax.plot(epochs, trainloss, label='Training loss')
    line1.set_dashes([2, 2, 10, 2]) 
    line1.set_dash_capstyle('round')
    line2, = ax.plot(epochs, testloss, dashes=[6, 2], label='Test loss')
    ax.legend(handlelength=4)
    plt.title(title)
    plt.xlabel('epochs')
    plt.ylabel('loss')
    plt.show()

graph_simple_plot(epochs, loss2, loss1, "Loss plot for using max as symmetric function")
print("NormalNet using max as symmetric function test loss at epoch 100: " + str(loss1[-1]))
graph_simple_plot(epochs, loss4, loss3, "Loss plot for using sum as symmetric function")
print("NormalNet using sum as symmetric function test loss at epoch 100: " + str(loss3[-1]))
graph_simple_plot(epochs, loss6, loss5, "Loss plot for using global features only and max pooling")
print("NormalNet using global feature only's test loss at epoch 100: " + str(loss6[-1]))
```
